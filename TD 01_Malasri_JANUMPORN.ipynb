#3. Pour combien
Implementer un script qui joue au jeu pour combien (1 à 10) avec l'utilisateur pour sortir d'une boucle

```
from random import randint
x = randint(1,10)
while(True) :
  n = input("Pour combien tu sors de cette boucle (1 à 10)? ")
  n = int(n)
  if n == x :
    print(f"Un...Deux...Trois: {n} ! Bravo, je te libère.")
    break
  elif n < x :
    print(f"Un...Deux...Trois: {n} ! Trop bas")
  elif n > x :
    print(f"Un...Deux...Trois {n} ! Trop haut")
```
