Pour trier une liste, on utilise la méthode **sort**. 


#EX02
Implementer une fonction qui :

1.   Prompte l'utilisateur d'input un nombre
2.   Retourne ce dernier si c'est bien un nombre
3.   Redemande un input sinon

````
def saisir() :
   x = input("Entrer un nombre: ")
   while not(x.isdigit()) :
     print(f"{x} n'est pas un nombre entier, veuillez re-essayer.")
     x = input("Entrer un nombre: ")
   return x
print("Valeur saisie: ",saisir())
````

#EX03

Implementer une fonction qui :

1.   Retourne l'extension d'un fichier donné par l'utilisateur (le nom complet du fichier en input)
  *   extension_fichier("test.pdf") -> pdf
  *   extension_fichier("test.png") -> png
  *   extension_fichier("test") -> raise une erreur qui dit que l'extension est inconnue




````
def getExtension() :
  while True :
    try :
      x = input("Entrer un nom de fichier complet: ")
      x = x.split(".")
      return x[1]
    except :
      print("L'extension est inconnue")
````

````
print("Extension fichier ->",getExtension())  
````
