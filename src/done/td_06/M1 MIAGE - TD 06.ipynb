```python
# Code here
# lire dataset 'tv_shows.csv' dans une variable tv_data
tv_data = pd.read_csv('tv_shows.csv', delimiter="|")
print(tv_data)

# Les 10 premi√®res lignes du dataset
print(tv_data.head(10))
```

```python
# Creer une colonne 'streaming_platform' qui contient le nom de la plateforme de streaming en se 
# basant sur les colonnes (hulu, prime, netflix,disney)

tv_data = pd.melt(
    tv_data, id_vars=['ID', 'Title', 'Year',
                      'Age', 'IMDb', 'Rotten Tomatoes', 'Type'],
    value_vars=["Netflix", "Hulu", "Prime Video", "Disney+"],
    var_name="streaming_platform",
    value_name="platform"
)
tv_data = tv_data[tv_data.platform == 1]

# drop les colonnes (hulu,prime, disney, netflix)

tv_data.drop('platform', axis=1, inplace=True)
print(tv_data.head(10))
```

```python
# afficher le nombre de shows par plateform de streaming

number_of_shows = tv_data.groupby(['streaming_platform'])[
    'streaming_platform'].count()
print(number_of_shows)
```

```python
# Afficher la note moyenne (imdb) par plateforme de streaming

tv_data[["Note", "td"]] = tv_data["IMDb"].str.split("/", expand=True)
tv_data = tv_data.astype({"Note": "float"})
tv_data.drop("td", axis=1, inplace=True)

avg_note_by_platform = tv_data.groupby(['streaming_platform'])["Note"].mean()
print(avg_note_by_platform)
```

```python
# Assigner a une variable top_shows, les meilleurs shows par plateforme de streaming et afficher cette derni√®re
top_shows = tv_data.groupby(['streaming_platform'])[
    'Note'].transform(max) == tv_data['Note']
print(tv_data[top_shows])
```
