```
# Importer la librairie pandas
import pandas as pd
import numpy as np

# lire dataset 'tv_shows.csv' dans une variable tv_data
file_path = "./tv_shows.csv"
tv_data = pd.read_csv(file_path, delimiter='|')

#Afficher les 10 premieres lignes du dataset
# print(tv_data[:10])
tv_data.head()
```

```
from pandas.core.groupby.groupby import T


print("--------------------------------------Sreaming_platform-------------------------------------------------------------")
# Cr√©er une colonne "sreaming_platform" qui contient le nom de la plateforme de streaming en se basant sur les colonnes (hulu, prime, disney, netflix)
conditions = [tv_data['Netflix']==1, tv_data['Disney+']==1, tv_data['Hulu']==1, tv_data['Prime Video']==1,]
choices = ['Netflix','Disney+','Hulu', 'Prime Video']
stream = tv_data.assign(sreaming_platform = np.select(conditions, choices, default=np.nan))
print(stream)
#solution optimale = methode melt, √† tester

print("---------------------------------------Drop Plateformes-----------------------------------------------------------")


# Drop les colonnes (hulu, prime, disney, netflix) en conservant streaming platform
drop_col_stream = tv_data.assign(sreaming_platform = np.select(conditions, choices, default=np.nan)).drop(['Netflix','Disney+','Hulu', 'Prime Video'], axis=1) # nan = not a number
print(drop_col_stream.head())
# Drop les colonnes (hulu, prime, disney, netflix) sans conserver streaming platform
tv_data.drop(['Netflix','Disney+','Hulu', 'Prime Video'], axis=1)

print("-----------------------------------------Nombre de show par plateforme------------------------------------------------------------")


# Afficher le nombre de shows par plateforme de streaming
n = tv_data['Netflix'].sum()
print ("Netflix show:",n)
d = tv_data['Disney+'].sum()
print ("Disney+ show:",d)
h = tv_data['Hulu'].sum()
print ("Hulu show:",h)
p = tv_data['Prime Video'].sum()
print ("Prime Video show:",p) 

print("----------------------------------------Nombre de show par plateforme 2----------------------------------------------------------")

print(stream['sreaming_platform'].value_counts())

# sum_show = tv_data.groupby(by=['Netflix','Disney+','Hulu', 'Prime Video']).sum().groupby(level=[0]).cumsum()
# print(sum_show)  ---------------- Pas compris

print("---------------------------------------Note moyenne par plateforme--------------------------------------------------------------------")

# Afficher la note moyenne (imdb) par plateforme de streaming

#Split la colonne IMDb pour r√©cup√©rer la note
drop_col_stream[['IMDb','/10']] = drop_col_stream['IMDb'].str.split('/',expand=True)
#print(drop_col_stream)

#Convertit le type de la colonne de chaine de caract√®re √† float
drop_col_stream['IMDb']=pd.to_numeric(drop_col_stream['IMDb'],errors='coerce').fillna(0)

#Moyenne avec groupBy 
show_mean = drop_col_stream.groupby('sreaming_platform')['IMDb'].mean()
print(show_mean)

#split data pour ne r√©cup√©rer que ce qu'il y a avant le slash pour calculer la moyenne

print("----------------------------------------Meilleurs show par plateforme------------------------------------------------------------------")

# Assigner a une variable top_shows, les meilleurs shows par plateforme de streaming et afficher cette derni√®re
# drop_col_stream.dropna(axis=0)

top_show = drop_col_stream.groupby('sreaming_platform')['IMDb'].apply(lambda x: x.idxmax()) #?
print(top_show)



#Questions: Pourquoi le dataframe s'affiche sans faire de print ?
#           Si plusieurs plateforme pour un show, comment g√©rer l'affichage ?
#           Modifier le fichier csv  ?
#           index  ?
```
